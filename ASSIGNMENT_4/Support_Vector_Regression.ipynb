1. IMPORTING LIBRARIES AND DATASET

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
dataset = pd.read_csv ('datasets_88705_204267_Real estate.csv')
X = dataset.iloc [ : ,2:-1].values
Y = dataset.iloc [ : ,-1].values
```

```
print (X)
```

```
print (Y)
```

2. SPLITTING INTO TRAINING AND TEST SET

```
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split (X, Y, test_size = 0.2, random_state = 0)
```

```
print (X_train)
```

```
print (Y_train)
```

```
print (X_test)
```

```
print (Y_test)
```

4. FEATURE SCALING

```
Y_train = Y_train.reshape (len (Y_train), 1)
```

```
Y_test = Y_test.reshape (len (Y_test), 1)
```

```
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler ()
sc_Y = StandardScaler ()
X_train = sc_X.fit_transform (X_train)
X_test = sc_X.transform (X_test)
Y_train = sc_Y.fit_transform (Y_train)
Y_test = sc_Y.transform (Y_test)
```

```
print (X_train)
```

```
print (Y_train)
```

5. SUPPORT VECTOR REGRESSION MODEL

5.1. TRAINING THE SVR MODEL

```
from sklearn.svm import SVR
regressor = SVR (kernel = 'rbf')
regressor.fit (X_train, Y_train)
```

```
Y_pred = sc_Y.inverse_transform (regressor.predict (X_test))
X_test = sc_X.inverse_transform (X_test)
Y_test = sc_Y.inverse_transform (Y_test)
```

```
print (Y_test)
```

5.2. VISUALISING TEST RESULTS

```
for i in range (5):
  plt.figure ()
  plt.scatter (X_test [ : ,i], Y_test, color = 'red')
  plt.scatter (X_test [ : ,i], Y_pred, color = 'blue')
  plt.title ('TARGET vs FEATURE ' + str (i+1))
  plt.xlabel ('FEATURE ' + str (i+1))
  plt.ylabel ('TARGET')
  plt.show ()
```
